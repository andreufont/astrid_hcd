# Compute linear 2pcf

```python
from nbodykit.lab import cosmology
import numpy as np
import matplotlib.pyplot as plt
```

```python
cosmo = cosmology.Cosmology(
    n_s=0.9667, 
    h=0.6774,
    A_s=2.142e-9,
    Omega0_b=0.0486,
    Omega0_cdm=0.3089-0.0486,
)
new_cosmo = cosmo.match(sigma8=0.816)
```

```python
Plin = cosmology.LinearPower(new_cosmo, redshift=3, transfer='CLASS')
```

```python
cf_lin = cosmology.CorrelationFunction(Plin)
```

```python
r = np.logspace(-1, 2.1, 1000)
plt.plot(r, r**2 * cf_lin(r))
```
